```python
### Simulating with beta.ppf 

ad_1 = []
ad_2 = []
ad_3 = []
ad_4 = []
ad_5 = []

for i in range(100000):
    ad_1.append(beta.ppf(np.random.rand(),53,949))
    ad_2.append(beta.ppf(np.random.rand(),39,963))
    ad_3.append(beta.ppf(np.random.rand(),52,950))
    ad_4.append(beta.ppf(np.random.rand(),46,956))
    ad_5.append(beta.ppf(np.random.rand(),26,976))
```

```python
### Simulating with beta.rvs 

ad_1 = []
ad_2 = []
ad_3 = []
ad_4 = []
ad_5 = []

for i in range(100000):
    ad_1.append(beta.rvs(53,949))
    ad_2.append(beta.rvs(39,963))
    ad_3.append(beta.rvs(52,950))
    ad_4.append(beta.rvs(46,956))
    ad_5.append(beta.rvs(26,976))
```

```python
df2 = pd.DataFrame(np.transpose([ad_1,ad_2,ad_3,ad_4,ad_5]), columns=['draws1','draws2','draws3','draws4','draws5'])
```

```python
### Creating ismax columns 
for i in range(1,6):
    df2[f'ismax{i}']= 0 
```

```python
### Filling ismax columns 
for i in range(100000):
    if df2['draws1'][i] == df2.loc[i].max():
        df2['ismax1'][i] = 1 
    if df2['draws2'][i] == df2.loc[i].max():
        df2['ismax2'][i] = 1 
    if df2['draws3'][i] == df2.loc[i].max():
        df2['ismax3'][i] = 1 
    if df2['draws4'][i] == df2.loc[i].max():
        df2['ismax4'][i] = 1 
    if df2['draws5'][i] == df2.loc[i].max():
        df2['ismax5'][i] = 1 
```
